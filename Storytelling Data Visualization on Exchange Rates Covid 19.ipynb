**Storytelling Data Visualization on Exchange Rates: Covid 19**

**Introduction**

This dataset describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union. The data source is the European Central Bank. Note that the dataset gets regular updates — we downloaded it on January 2021. Data can be found on kaggle: https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020

The aim of this project is to show the how the euro-dollar rate was affected during Covid 19. 

**Introducing Dataset**

```python
import pandas as pd #Import pandas and reading the data
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head() #Displaying first five rows
```

```python
exchange_rates.tail() #Displaying last five rows
```

```python
exchange_rates.info() #checking the data type and the number of rows and columns
```

There is 40 columns and 5699 rows, there is no null values as expected. Only two colums are floats they rest are objects which is unexpected.

**Data Cleaning**

```python
#Renaming the 'US Dollar' and 'Period\\Unit:' colims to 'US_Dollar' and 'Time' respectively
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar','Period\\Unit:': 'Time'}, inplace=True) 

#Converting the 'Time' column to a datetime data type
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

#Sorting the 'Time' column in ascending order
exchange_rates.sort_values('Time', inplace=True)

#Reset the index and drop the initial index
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']]

#Checking the counts of unique values in the 'US_dollar' column
euro_to_dollar['US_dollar'].value_counts()
```

There is 62 counts of '-' appearing the 'US_dollar' column we will remove this as it provides no information

```python
#Removing all rows where '-' appears in the US_dollar column
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']

#Converting the euro_to_dollar column to a float date type
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

#Checking the data type of the two columns
euro_to_dollar.info()
```

**Rolling Mean**

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show() #Generating a line graph to show the relationship between the Euro and Dollar across time
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

#Plotting a rolling window of 7, 30, 50, 100 and 365 days to see which is best
for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]): 
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() 
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python

euro_to_dollar['Time']= pd.to_datetime(euro_to_dollar['Time'])
```

```python
euro_to_dollar.info()
```

```python
#Using only data from 2020-2021 to see effect of Covid19
covid_19 = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2020
                   ) & (euro_to_dollar['Time'].dt.year <= 2021)]
```

**Effects of Coronavirus on the Euro-Dollar rate**

```python
# Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

# Adding the plot
fig, ax= plt.subplots(figsize=(10,6))
ax.plot(covid_19['Time'],
        covid_19['rolling_mean'],
        linewidth=2, color='#b00b1e')

ax.tick_params(axis='x', colors='grey')
ax.tick_params(axis='y', colors='grey')


#Removing boarders
for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)

#Adding text to visualisation
ax.text(0.34, 0.08, '1.085', alpha=0.5, transform=ax.transAxes)
ax.text(0.45, 0.09, 'Lockdown Starts to Ease in Europe', alpha=0.5, transform=ax.transAxes)
ax.text(0.17, 0.2, '1.09', alpha=0.5, transform=ax.transAxes)
ax.text(0.1, 0.26, 'Coronavirus Breakout in Europe', alpha=0.5, transform=ax.transAxes)
ax.text(0.98, 0.95, '1.215', alpha=0.5, transform=ax.transAxes)
ax.text(0.7, 0.6, 'Vaccine Rollout', alpha=0.5, transform=ax.transAxes)

#Adding heading
ax.text(-0.1, 1.1, "Euro-USD rate Hit Lows of 1.085 During the Pandemic",
        weight='bold',transform=ax.transAxes)
#Adding sub-heading
ax.text(-0.1, 1.05, 'Euro-USD exchange rates between 2020 and 2021',
        size=12, transform=ax.transAxes)
#Adding a signature
ax.text(-0.1, -0.2, '©Zeinab' + ' '*94 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=15, transform=ax.transAxes)
plt.show()
```

**Conclusion**

As the pandemic starts to enter Europe in March, we can see a dip in the Eur-USD rate, hitting a low in May. Lockdown starts to ease throughout many European states in June 2020, the increased certanity causes the rate to increase again. Europe goes through a second wave in October causing a slight dip in the exchange rate. Mass vaccination rollout schemes in November, indicating the economy going back to normal has further strengthed the rate.
